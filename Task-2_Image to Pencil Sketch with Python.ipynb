## Importing The Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

## Reading the Img

```python
original_img = cv2.imread('480x640_Beautiful_Lake(wapkingin).jpg')
# original_img = cv2.imread('testing_image.jpg')
```

```python
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(original_img))
```

## Converting the Original Image into Grayscale 

```python
gray_image = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

## Inverting the Grayscale image

```python
inv_gray_image = 255 - gray_image
```

```python
inv_gray_image
```

```python
display(Image.fromarray(inv_gray_image))
```

## Mixing the grayscale image and inverted blur image as well

```python
inv_blur_img = 255 - gray_image

def dodge(gray_image, inv_blur_img):
    return cv2.divide(gray_image, inv_blur_img, scale = 300)
```

```python
pencil_sketch = dodge(gray_image, inv_blur_img)
display(Image.fromarray(pencil_sketch))
```
